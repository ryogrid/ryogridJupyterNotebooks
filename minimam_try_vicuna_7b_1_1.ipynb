```python
from transformers import AutoTokenizer, AutoModelForCausalLM, T5Tokenizer
tokenizer = AutoTokenizer.from_pretrained("eachadea/vicuna-7b-1.1")
```

```python
model = AutoModelForCausalLM.from_pretrained("eachadea/vicuna-7b-1.1")
```

```python
def generate_prompt(instruction, input=None):
    if input:
        return f"""Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

# Instruction:
{instruction}

# Input:
{input}

# Response:
"""
    else:
        return f"""Below is an instruction that describes a task. Write a response that appropriately completes the request.

# Instruction:
{instruction}

# Response:
"""
```

```python
# 推論
prompt=generate_prompt("answer the question", "何故雨が降るのか?")
input = tokenizer.encode(prompt, return_tensors="pt")
output = model.generate(input, do_sample=True, max_length=200, num_return_sequences=1)
ans=(tokenizer.batch_decode(output))

print(ans[0].replace(".\\n\\n2.","\\n"))
```
